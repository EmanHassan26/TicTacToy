```python
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
from random import randint
 
root = Tk()
root.title("TIC TAC TOY ")
root.config(background="pink")

activep = 1
p1 = [] #what player one select
p2 = [] #what player two select
winner = -1
empty = []
#////////////////////////////////////////////////////////////////
def setout( id , x )  :
    if id == 1:
        b1.config(text=x , state='disable')
    elif id == 2:
        b2.config(text=x , state='disable')    
    elif id == 3:
        b3.config(text=x , state='disable')
    elif id == 4:
        b4.config(text=x , state='disable')
    elif id == 5:
        b5.config(text=x , state='disable')
    elif id == 6:
        b6.config(text=x , state='disable')
    elif id == 7:
        b7.config(text=x , state='disable')
    elif id == 8:
        b8.config(text=x , state='disable')
    elif id == 9:
        b9.config(text=x , state='disable')
        
        
#////////////////////////////////////////////////////////////////   
def biclic(id):
    global activep
    global p1
    global p2
    global empty
    if activep == 1 :
        setout( id , "X" )
        p1.append(id)
        empty.append(id)
        activep = 2
        print(f" p1 : {p1}")
        #Autoplay()
    elif activep == 2 :
        setout( id , "O" )
        p2.append(id)
        empty.append(id)
        activep = 1
        print(f" p2 : {p2}")    
    checkwinner()
    
#////////////////////////////////////////////////////////////////    
def checkwinner ():
    winner=-1
    if( (1 in p1) and (2 in p1) and (3 in p1)) :
        winner = 1
    elif( (1 in p2) and (2 in p2) and (3 in p2)) :
        winner = 2
    elif( (4 in p1) and (5 in p1) and (6 in p1)) :
        winner = 1
    elif( (4 in p2) and (5 in p2) and (6 in p2)) :
        winner = 2
    elif( (7 in p1) and (8 in p1) and (9 in p1)) :
        winner = 1   
    elif( (7 in p2) and (8 in p2) and (9 in p2)) :
        winner = 2    
    elif( (1 in p1) and (4 in p1) and (7 in p1)) :
        winner = 1   
    elif( (1 in p2) and (4 in p2) and (7 in p2)) :
        winner = 2  
    elif( (2 in p1) and (5 in p1) and (8 in p1)) :
        winner = 1    
    elif( (2 in p2) and (5 in p2) and (8 in p2)) :
        winner = 2 
    elif( (3 in p1) and (6 in p1) and (9 in p1)) :
        winner = 1        
    elif( (3 in p2) and (6 in p2) and (9 in p2)) :
        winner = 2    
    elif( (1 in p1) and (5 in p1) and (9 in p1)) :
        winner = 1 
    elif( (1 in p2) and (5 in p2) and (9 in p2)) :
        winner = 2     
    elif( (3 in p1) and (5 in p1) and (7 in p1)) :
        winner = 1   
    elif( (3 in p2) and (5 in p2) and (7 in p2)) :
        winner = 2   
        
        
    if  winner == 1 :
        messagebox.showinfo(title="Congratulations !" , message="Player1 Won !")
    elif  winner == 2 :
        messagebox.showinfo(title="Congratulations !" , message="Player2 Won !")
    elif len(empty)==9 and winner != 1 and winner != 2 :
        messagebox.showinfo(title="Congratulations !" , message="Game Over")    
#////////////////////////////////////////////////////////////////  
# def Autoplay():
#     global p1
#     global p2
#     empty=[]
#     for x in range(9):
#         if not((x+1 in p1) or (x+1 in p2)):
#             empty.append(x+1)
#     rand=randint(0 , len(empty)-1)        
#     biclic(empty[rand])    

#//////////////////////////////////////////////////////////////// 

b1=Button(text='' , font="30",fg="black")
b1.grid(row=0 , column=0 , sticky='snew', ipadx = 50  ,ipady = 50 )
b1.config(command= lambda : biclic(1))


b2=Button(text='' , font="30",fg="black")
b2.grid(row=0 , column=1 , sticky='snew', ipadx = 50  ,ipady = 50 )
b2.config(command= lambda : biclic(2))


b3=Button(text='' , font="30",fg="black")
b3.grid(row=0 , column=2 , sticky='snew', ipadx = 50  ,ipady = 50 )
b3.config(command= lambda : biclic(3))


b4=Button(text='' , font="30",fg="black")
b4.grid(row=1 , column=0 , sticky='snew', ipadx = 50  ,ipady = 50 )
b4.config(command= lambda : biclic(4))


b5=Button(text='' , font="30",fg="black")
b5.grid(row=1 , column=1 , sticky='snew', ipadx = 50  ,ipady = 50 )
b5.config(command= lambda : biclic(5))


b6=Button(text='' , font="30",fg="black")
b6.grid(row=1 , column=2 , sticky='snew', ipadx = 50  ,ipady = 50 )
b6.config(command= lambda : biclic(6))


b7=Button(text='' , font="30",fg="black")
b7.grid(row=2 , column=0 , sticky='snew', ipadx = 50  ,ipady = 50 )
b7.config(command= lambda : biclic(7))


b8=Button(text='', font="30",fg="black")
b8.grid(row=2 , column=1 , sticky='snew', ipadx = 50  ,ipady = 50 )
b8.config(command= lambda : biclic(8))


b9=Button(text='' , font="30",fg="black")
b9.grid(row=2 , column=2 , sticky='snew', ipadx = 50  ,ipady = 50 )
b9.config(command= lambda : biclic(9))


root.mainloop()
```
